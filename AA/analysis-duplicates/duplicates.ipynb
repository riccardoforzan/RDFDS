```python
import json
```

```python
json_datasets_file = "datasets.json"
```

```python
f = open(json_datasets_file)
data = json.load(f, strict=False)
f.close()
```

```python
datasets = sorted(data["datasets"], key=lambda d: int(d['dataset_id']))
```

```python
link_dataset = dict()

for entry in datasets:
    dataset_id = entry["dataset_id"]

    distinct_links = list(set(entry["download"]))

    for download_link in distinct_links:
        if download_link not in link_dataset.keys():
            link_dataset[download_link] = list()

        l = link_dataset[download_link]
        l.append(dataset_id)
```

```python
from enum import Enum


class Compare(Enum):
    DUPLICATE = 0
    LINK_IN_COMMON = 1


def compare_datasets(datasets: list, ids: list):
    # key = url , value = dataset IDs that downloads the same URLs
    d = dict()

    for id in ids:
        entry = next(filter(lambda e: e["dataset_id"] == id, datasets))

        # use as key a sorted string containing the downloaded URLs
        key = str(sorted(entry["download"]))
        if key not in d:
            d[key] = list()

        d[key].append(id)

    return (Compare.DUPLICATE if len(d.items()) == 1 else Compare.LINK_IN_COMMON, d)
```

```python
to_check = set()

for url, datasets_list in link_dataset.items():
    if len(datasets_list) > 1:
        to_check.add(tuple(sorted(datasets_list)))
```

```python
duplicati = list()
link_in_common = list()

for t in to_check:
    ids = list(t)
    res = compare_datasets(datasets, ids)
    
    if res[0] == Compare.DUPLICATE:
        duplicati.append(res[1])

    if res[0] == Compare.LINK_IN_COMMON:
        link_in_common.append(res[1])
```

```python
counter = 0
for d in duplicati:
    counter += len(d[next(iter(d))])

counter -= len(duplicati)

print(f"DUPLICATI: {counter}")
print(json.dumps(duplicati, indent=2))
```
