```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tdf = pd.read_csv("features.csv")
```

### Correlation analysis

```python
correlation = tdf.drop(["dataset_id", "relevance"], axis=1).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="crest")
plt.title("Correlation between extracted features")
plt.show()
```

```python
tmp = tdf.drop("dataset_id", axis=1).corr()["relevance"]
correlation = tmp.drop("relevance")

plt.bar(correlation.index, correlation)
plt.xlabel("Columns")
plt.ylabel("Correlation with Relevance")
plt.title("Correlation between Relevance and all the extracted features")
plt.xticks(rotation=90)
plt.show()
```

### Analyzing features / relevance relationship without non relevant datasets from the dataframe

```python
we_df = tdf[tdf['relevance'] != 0]
we_df.head()
```

```python
correlation = we_df.drop(["dataset_id", "relevance"], axis=1).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="crest")
plt.title("Correlation between extracted features")
plt.show()
```

```python
correlation = we_df.corr()['relevance'].drop(['relevance','dataset_id'])

plt.bar(correlation.index, correlation)
plt.xlabel('Columns')
plt.ylabel('Correlation with Relevance')
plt.title('Correlation between Relevance and all the extracted features')
plt.xticks(rotation=90)
plt.show()
```
